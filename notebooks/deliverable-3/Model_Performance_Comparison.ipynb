```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a dictionary with model performance results
model_results = {
    "Gradient Boosting": {"accuracy": 0.8142, "f1_macro": 0.8103},
    "XGradient Boosting": {"accuracy": 0.7699, "f1_macro": 0.7708},
    "Stacking": {"accuracy": 0.7699, "f1_macro": 0.7684},
    "Random Forest": {"accuracy": 0.7611, "f1_macro": 0.7581},
    "Ada Boost": {"accuracy": 0.7345, "f1_macro": 0.7330},
    "Bagging": {"accuracy": 0.7345, "f1_macro": 0.7348},
    "KNN": {"accuracy": 0.7522, "f1_macro": 0.7505},
    "NN": {"accuracy": 0.9813, "f1_macro": 0.9815}
}

# Convert the dictionary into a DataFrame for comparison
results_df = pd.DataFrame([
    {"Model": name, "Accuracy": result["accuracy"], "F1 Macro": result["f1_macro"]}
    for name, result in model_results.items()
])

# Sort the results by Accuracy in ascending order
results_df = results_df.sort_values(by="Accuracy", ascending=True)

# Display the results
print("Model Comparison:\n")
print(results_df)

# Create a bar chart to visualize model performance
results_df.set_index("Model")[["Accuracy", "F1 Macro"]].plot(kind='bar', figsize=(10, 6), color=["blue", "orange"])
plt.title("Model Performance Comparison")
plt.ylabel("Score")
plt.ylim(0, 1)  
plt.xticks(rotation=45, ha='right')  
plt.legend(loc='lower right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
